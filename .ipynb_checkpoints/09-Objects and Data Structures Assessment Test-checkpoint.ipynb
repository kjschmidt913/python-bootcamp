## Dictionaries

Using keys and indexing, grab the 'hello' from the following dictionaries:

```python
d = {'simple_key':'hello'}
# Grab 'hello'
d['simple_key']
```

```python
d = {'k1':{'k2':'hello'}}
# Grab 'hello'
d['k1']['k2']
```

```python
# Getting a little tricker
d = {'k1':[{'nest_key':['this is deep',['hello']]}]}

#Grab hello
d['k1'][0]['nest_key'][1][0]
```
